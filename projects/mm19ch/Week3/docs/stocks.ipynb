```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
start = datetime.datetime(2010, 1, 1)
end   = datetime.datetime(2021, 5, 30)
```

```python
aapl = web.DataReader("aapl", 'yahoo', start, end)
msft = web.DataReader("msft", 'yahoo', start, end)
googl = web.DataReader("googl", 'yahoo', start, end)
```

```python
aapl
msft
googl
```

```python
msft['Open'].plot()
plt.show()

aapl['Open'].plot()
plt.show()

googl['Open'].plot()
plt.show()
```

```python
aapl['Open'].plot(label = 'AAPL Open price', figsize=(15,5))
msft['Open'].plot(label = 'MSFT Open price')
googl['Open'].plot(label = 'GOOGL Open price')
plt.legend()
plt.title('Comparison')
plt.show()
```

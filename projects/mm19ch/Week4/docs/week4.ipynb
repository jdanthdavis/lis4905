```python
def data_analysis():
    print("Print indexes")
    print(df.index)
    print("\Print columns")
    print(df.columns)
    print("\nPrint data frame")
    print(df.head(5))
    print(df.tail(5))
    print("\nPrint type")
    print(type(df))
    print("\nPrint data attribute type")
    print(df.dtypes)
    print("\nPrint values in array format")
    print(df.values)
    print("\nPrint summary stats")
    print(df.info(verbose=True))
    print("\nNumber of rows")
    print(df.shape[0])
    print("\nNumber of columns")
    print(df.shape[1])
    print("\nNumber of rows and columns")
    print(df.shape)
    print("\nNumber of all elements")
    print(df.size)
    print("\nDisplay first and last 5 species")
    print(df.iloc[ :, 5])
    print("\nDisplay all unique species")
    print(df.spec_id.unique())
    print("\nDisplay total number of species")
    print(df['spec_id'].count())
    print("\nDisplay number of unique species")
    print(df['spec_id'].nunique()+1)
    print("\nDisplay number of unique species excluding not identified")
    print(df['spec_id'].nunique())
    print("\nDisplay stats of animal weight")
    weight = df['weight'].describe()
    print(weight)
    print("\nDisplay weight metricts")
    print("min ", df['weight'].min())
    print("max ", df['weight'].max())
    print("mean ", df['weight'].mean())
    print("std ", df['weight'].std())
    print("count ", df['weight'].count())
    print("\nDisplay mean weight")
    print(round(df['weight'].mean(),2))
    print("\nDisplay number of records per species")
    print(df.spec_id.value_counts())
    print("\nDisplay number of records per specific species")
    print(df[df['spec_id'] == 'PF'].shape[0])
    print("\nDisplay chart of records per species")
    chart = df['spec_id'].value_counts().reset_index()
    chart.columns = ['Species', 'total per species']
    chart = chart.sort_values(by = 'Species', ascending = True)
    chart.plot.bar(x = 'Species', y = 'total per species')
    plt.show()
    print("\nDisplay chart of how many unique species are recorded from each location")
    chart_loc = df['loc_id'].value_counts().reset_index()
    chart_loc.columns = ['Location', 'total per location']
    chart_loc = chart_loc.sort_values(by = 'Location', ascending = True)
    chart_loc.plot.bar(x = 'Location', y = 'total per location')
    plt.show()
```

```python
data_analysis()
```

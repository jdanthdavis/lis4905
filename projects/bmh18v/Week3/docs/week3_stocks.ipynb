```python
def displayGraph1():
        start = datetime.datetime(2010, 1, 1)
        end = datetime.datetime.now()
        df = pdr.DataReader("AAPL", "yahoo", start, end)
        style.use('ggplot')
        df['High'].plot()
        df['Adj Close'].plot()
        plt.title("Apple Stocks - ggplot")
        plt.legend()
        plt.show()
```

```python
import pandas_datareader.data as pdr
import datetime
import matplotlib.pyplot as mpl
```

```python
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2021, 6, 3)
```

```python
msft = pdr.DataReader("msft", "yahoo", start, end)
```

```python
ada = pdr.DataReader( "ADA-USD","yahoo", start, end)
```

```python
doge = pdr.DataReader( "DOGE-USD", "yahoo", start, end)
```

```python
mpl.style.use('ggplot')
msft['High'].plot()
msft['Adj Close'].plot()
mpl.title("Microsoft Stocks")
mpl.legend()
mpl.show()
```

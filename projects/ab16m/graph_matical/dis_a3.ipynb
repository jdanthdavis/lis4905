```python
def data_analysis():
    start = datetime.datetime(2010, 1, 1)
    end = datetime.datetime(2021, 1, 1)
    df = pdr.DataReader("BTC-USD", "yahoo", start, end)
    num = len(df)
    print(f"\nPrint number of records:\n{num}")
    print("\nPrint columns:")
    print(df.columns)
    print("\nPrint Data Frame:")
    print('\nNote: pandas displays "max_rows" number of rows (default, first and last 5).')
    print(df)
    print("\nPrint first 8 rows:")
    print(df.iloc[0:8])
    print("\nPrint last 6 rows:")
    print(df.tail(6))
```

```python
def printGraph1():
    start = datetime.datetime(2010, 1, 1)
    end = datetime.datetime(2021, 1, 1)
    df = pdr.DataReader("BTC-USD", "yahoo", start, end)
    style.use('ggplot')
    df['High'].plot()
    df['Adj Close'].plot()
    plt.title("Bitcoin Stocks - ggplot")
    plt.legend()
    plt.show()
    
```

```python
def printGraph2():
    start = datetime.datetime(2010, 1, 1)
    end = datetime.datetime(2021, 1, 1)
    df = pdr.DataReader("ETSY", "yahoo", start, end)
    style.use('classic')
    df['High'].plot()
    df['Adj Close'].plot()
    plt.title("Etsy Stocks - classic")
    plt.legend()
    plt.show()
```

```python
def printGraph3():
    start = datetime.datetime(2010, 1, 1)
    end = datetime.datetime(2021, 1, 1)
    df = pdr.DataReader("TSLA", "yahoo", start, end)
    style.use('fivethirtyeight')
    df['High'].plot()
    df['Adj Close'].plot()
    plt.title("Tesla Stocks - fivethirtyeight")
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_requirements():
    print('Program Requirements')
    print('1. Automate data manipulation process')
    print('2. Easily reproduce analysis on different datasets using similar format -- saves time!')
    print('3. Import necessary libraries.')
    print('4. Research how to install any missing packages, if necessary.')
    print('5. Create at least three functions that are called by the program:')
    print('\ta. main(): calls at least two other functions.')
    print('\tb. get_requirements(): displays the program requirements.')
    print('\tc. data_analysis_2(): displays the following data.')
```

```python
def data_analysis_2(df):
    print(" ")
    print("1. Print indexes:")
    print(df.index)
    print(" ")
    print("2. Print columns:")
    print(df.columns)
    print(" ")
    print("3. Print data frame:")
    print(df)
    print(" ")
    print("4. Print type:")
    print(type(df))
    print(" ")
    print("5. Print attribute data types--object type:")
    print(df.dtypes)
    print(" ")
    print("6. Print values (truncated list):")
    print(df.values)
    print(" ")
    print("7. Print index of all Dataframe column names:")
    print(df.columns)
    print(" ")
    print("Summary Statistics")
    print(" ")
    print("8. Print Dataframe info:")
    print(df.info())
    print(" ")
    print("9. Print *only* number of Dataframe rows:")
    print(df.shape[0])
    print(" ")
    print("10. Print *only* number of Dataframe columns:")
    print(df.shape[1])
    print(" ")
    print("11. Print number of Dataframe rows and columns:")
    print(df.shape)
    print(" ")
    print("12. Print number of DataFrame elements:")
    print(df.shape[0] * df.shape[1])
    print(" ")
    print("13. Display first and last 5 species:")
    print(df['spec_id'])
    print(" ")
    print("14. Display all *unique* species:")
    print(df['spec_id'].unique())
    print(" ")
    print("15. Display total number of species caught")
    print(len(df['spec_id']))
    print(" ")
    print("16. Display the number of *unique* species caught:")
    print(df['spec_id'].nunique()+1)
    print(" ")
    print("17. Display the number of *unique* species caught, excluding species not identifiied:")
    print(df['spec_id'].nunique())
    print(" ")
    print("18. Display summary statistics of animal weights:")
    x = df['weight'].describe()
    print(x)
    print(" ")
    print("19. Display individual aggregate weight metrics:")
    print("min:" , df['weight'].min())
    print("max:" , df['weight'].max())
    print("mean:" , df['weight'].mean())
    print("std:" , df['weight'].std())
    print("count:" , df['weight'].count())
    print(" ")
    print("20. Display mean weight, rounded to two decimal places:")
    print("mean:" , round(df['weight'].mean(),2))
    print(" ")
    print("21. Groupby one column (sex), and display mean of only *one* specific column in group(weight):")
    print(df.groupby('sex')['weight'].mean())
    print(" ")
    print('22. Display number of records per species:')
    print(df['spec_id'].value_counts().sort_index(ascending=True))
    print(" ")
    print('23. Display number of records per specific species (PF):')
    print(df[df['spec_id'] == 'PF'].shape[0])
    print(" ")
    print("24. Display bar chart of number of records per species, include legend:")
    df_graph = df['spec_id'].value_counts().reset_index()
    df_graph.columns = ['Species', 'total per species']
    df_graph =  df_graph.sort_values(by = 'Species' ,ascending = True)
    df_graph.plot.bar(x = 'Species', y = 'total per species')
    plt.show()
    print("25. Display bar chart of how many unique species recorded from each location, excluding missing values, include legend:")
    df_graph1 = df['loc_id'].value_counts().reset_index()
    df_graph1.columns = ['Location','total unique species per location']
    df_graph1 = df_graph1.sort_values(by = 'Location' , ascending = True)
    df_graph1.plot.bar(x = 'Location' , y = 'total unique species per location')
```

```python
def main():
    df_as = pd.read_csv('animal_survey.csv')
    get_requirements()
    data_analysis_2(df_as)
```

```python
main()
```

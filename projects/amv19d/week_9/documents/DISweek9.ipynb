```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('imdb-movie-data.csv')
```

```python
def watch_rating(x):
    if x >= 8.5:
        return("yes")
    else:
        return("no")
```

```python
def analyze_df(df):
    print("26. Frequency of all values (value_counts()) in a column (here, 'genre'), return 1st 10:")
    print(df['Genre'].value_counts()[0:10])
    print(" ")
    print("27. Relationships between continuous variables using .corr() method: \n Pos. numbers: pos. correlation. Neg. numbers: inverse correlation. 1.0 indicates perfect correlation. \n Correlation of interest: votes and revenue millions.")
    print(" ")
    print(df.corr())
    print(" ")
    print("28.Conditional Selections.")
    print("28a. Display oldest and newest release years for movies in dataset: ")
    print("Oldest: ", df["Year"].max(),", Newest: ", df["Year"].min(),sep="")
    print("28b. Filter by one director, return 3 records:")
    print(df[df["Director"]=="Clint Eastwood"].head(3))
    print("28c. Filter by rating (>=8.0), return 3 records: ")
    print(df[df["Rating"]>=8.0].head(3))
    print(" ")
    print("28d. Filter by two directors (using 'or'). Logical operators | for 'or'  and &  for 'and', return 1st 5 records")
    print(df[(df["Director"]=="Clint Eastwood")|(df["Director"]=="James Gunn")].head())
    print("28e. Filter by same two directors (using 'isin'), return 1st 5 records: ")
    print(df[df["Director"].isin(["Clint Eastwood","James Gunn"])].head())
    print(" ")
    print("28f. Filter by all movies released between 2005 and 2010, inclusive, \n having a rating abot 8.5:")
    print(df[(df["Year"] >= 2005) & (df["Year"]<= 2010)&(df["Rating"]>8.5)])
    print("28g. Filter by all movies released btween 2005 and 2010, inclusive, \n having rating above 8.0, and made aboce or equal to 25th percentile in revenue:")
    print(df[(df["Year"] >= 2005) & (df["Year"]<= 2010)&(df["Rating"]>8.0)& (df['Revenue (Millions)'] > df['Revenue (Millions)'].quantile(.75))])
    print("")
    print("29. Create function, watch _rating(), returns 'yes' or 'no' based upon 8.5 or greater rating. \n Use Python's apply() function to pass 'rating' attribute to function, \n and assign return values to new attribute 'watch_movie', display 1st 10 rows.")
    df["watch_movie"]=df["Rating"].apply(lambda x: watch_rating(x))
    print(df.head(10))
    print(" ")
    print("30.Categorical variables: bar charts and boxplots. \n Continuous variables: Histograms, Scatterplots, Line graphs, and Boxplot. \n Create following plots: scatter, histogram, two boxplots, and one heatmap.")
    sns.set_theme(style="darkgrid")
    fig, ax = plt.subplots()
    ax.scatter(df["Rating"],df["Revenue (Millions)"],color="orange")
    ax.set_xlabel('rating')
    ax.set_ylabel('revenue_millions')
    ax.set_title("Revenue(millions) VS Rating", fontsize = 20)
    sns.set_theme(style="white")
    fig, ax = plt.subplots()
    ax.hist(df["Rating"], color = 'hotpink')
    ax.set_ylabel('Frequency')
    ax.set_title('Rating')
    sns.set_theme(style = 'dark')
    fig, ax = plt.subplots()
    ax.boxplot(df['Rating'])
    plt.xticks([])
    ax.set_xlabel('rating')
    sns.set_theme(style="whitegrid")
   
    

        
        
```

```python
analyze_df(df)
```

```python
def analyze_df(df):
    print(" ")
    print("***DataFrame composed of three components: index, columns, and data. Data also known as values.***")
    print(" ")
    print("1. Print Index:")
    print(df.index)
    print(" ")
    print("2. Print Columns:")
    print(df.columns)
    print(" ")
    print("3. Print Dataframe:")
    print(df)
    print(" ")
    print("4. Print type:")
    print(type(df))
    print(" ")
    print("5. Print attribute data types:")
    print(df.dtypes)
    print(" ")
    print("6. Print values (truncated list), in array format:")
    print(df.values)
    print(" ")
    print("7. Print first 5 records in non-array format:")
    print(df.columns)
    print(" ")
    print("8. Print index of all Dataframe column names:")
    print(df.columns)
    print(" ")
    print("9. Print Dataframe info(i.e summary, similar to 'describe tablename;' in MySQL ): ")
    print(df.describe)
    print(" ")
    print("10. Print *only* number of Dataframe rows: ")
    print(df.shape[0])
    print(" ")
    print("11. Print *only* number of DataFrame columns:")
    print(df.shape[1])
    print(" ")
    print("12. Print number of DataFrame rows and columns:")
    print(df.shape)
    print(" ")
    print("13. Print number of Dataframe elements:")
    print(df.shape[0] * df.shape[1])
    print(" ")
    print("14. Split into input(x) and output (y) variables: Indicate how x/y variables are split. ")
    x = df.iloc[:, 0:8]
    y = df.iloc[:, 8]
    print("\n\ta. Display input (x) variables: ")
    print(x)
    
    print("\n\tb. Display output (y) variable (class): diabetic (1), non-diabetic (0):")
    print(y)
    
    print("15. Define, compile, fit, and evaluate keras model on dataset: ")
    model = Sequential()
    model.add(Dense(12, input_dim=8, activation='relu'))
    model.add(Dense(8, activation = 'relu'))
    model.add(Dense(1, activation = 'sigmoid'))
    
    model.compile(loss = 'binary_crossentropy',
                  optimizer = 'adam', metrics = ['accuracy'])
    model.fit(x, y, epochs = 150, batch_size=10, verbose = 0)
    _, accuracy = model.evaluate(x, y, verbose = 0)
    print('Accuracy %.2f' % (accuracy * 100))
    
    predictions = model.predict_classes(x)
    
    for i in range(1-10):
        print('%s => %d (expected %d)' % (x[i].tolist(), predictions[i], y[i]))
    
    
        
```

```python
def main():
    df = pd.read_csv('pima-indians-diabetes.csv')
    get_requirements()
    analyze_df(df)
```
